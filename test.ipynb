```python
from time import strftime
def get_timestamp(fmt: str = "%Y-%m-%d_%H:%M:%S") -> str:
    """Return the current local time formatted using `strftime` from `time`.

    Example: `20260101_153012`
    """
    return strftime(fmt)
get_timestamp()
```

```python
import pyautogui as pg

real_size = (1920, 1080)

size = pg.size()

diff_x = size[0] / real_size[0]
diff_y = size[1] / real_size[1]

scaled_x = round(636 * (size[0] / real_size[0]))
scaled_y = round(592 * (size[1] / real_size[1]))

print((scaled_x, scaled_y))
print(diff_x, diff_y)
```
