```python
import cv2
import pytesseract
from pytesseract import Output
from collections import defaultdict

# ─────────────────────────────────────────────
# CONFIG (Windows only – uncomment if needed)
# pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"
# ─────────────────────────────────────────────

IMAGE_PATH = "CMISAOP_1.png"  # <-- change to your image

# Load image
img = cv2.imread(IMAGE_PATH)
if img is None:
    raise FileNotFoundError("Image not found")

rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Run OCR
data = pytesseract.image_to_data(rgb, output_type=Output.DICT)

# Dictionary to group words by sentence (line)
lines = defaultdict(list)

n = len(data["text"])

for i in range(n):
    text = data["text"][i]

    # ✅ Ignore empty text & single spaces
    if not text or not text.strip():
        continue

    # Line identifier (block, paragraph, line)
    line_id = (
        data["block_num"][i],
        data["par_num"][i],
        data["line_num"][i],
    )

    lines[line_id].append(i)

# ─────────────────────────────────────────────
# Process each sentence (line)
# ─────────────────────────────────────────────
print("\nDetected sentences:\n")

for line_key, indices in lines.items():
    words = []
    xs, ys, xe, ye = [], [], [], []

    for i in indices:
        word = data["text"][i].strip()
        words.append(word)

        x = data["left"][i]
        y = data["top"][i]
        w = data["width"][i]
        h = data["height"][i]

        xs.append(x)
        ys.append(y)
        xe.append(x + w)
        ye.append(y + h)

    # Build sentence
    sentence = " ".join(words)

    # Bounding box for entire sentence
    x1, y1 = min(xs), min(ys)
    x2, y2 = max(xe), max(ye)

    # Draw box
    cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)

    # Draw coordinates
    label = f"({x1},{y1})"
    cv2.putText(img, label, (x1, y1 - 5),
                cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 1)

    # Print sentence
    print(f"Sentence: {sentence}")
    print(f"Position : x={x1}, y={y1}\n")

# Show result
cv2.imshow("Sentence-level OCR", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import pytesseract
from pytesseract import Output

# ─────────────────────────────────────────────
# CONFIG (Windows only)
# pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"
# ─────────────────────────────────────────────

IMAGE_PATH = "CMISAOP_1.png"

# Load image
img = cv2.imread(IMAGE_PATH)
if img is None:
    raise FileNotFoundError("Image not found")

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# ───────────── OCR (LABEL DETECTION) ─────────────
ocr_data = pytesseract.image_to_data(gray, output_type=Output.DICT)

labels = []

n = len(ocr_data["text"])
for i in range(n):
    text = ocr_data["text"][i].strip()

    # Ignore noise / single chars
    if len(text) < 2:
        continue

    x = ocr_data["left"][i]
    y = ocr_data["top"][i]
    w = ocr_data["width"][i]
    h = ocr_data["height"][i]

    labels.append({
        "text": text,
        "x": x,
        "y": y,
        "w": w,
        "h": h
    })

# ───────────── INPUT BOX DETECTION ─────────────
# Edge detection
edges = cv2.Canny(gray, 50, 150)

# Find contours
contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

input_boxes = []

for cnt in contours:
    x, y, w, h = cv2.boundingRect(cnt)

    # Heuristic for input boxes
    if w > 120 and 20 < h < 50:
        input_boxes.append((x, y, w, h))

# ───────────── MATCH LABEL → INPUT BOX ─────────────
fields = []

for label in labels:
    lx, ly = label["x"], label["y"]
    best_box = None
    min_dist = 9999

    for (bx, by, bw, bh) in input_boxes:
        # Box should be to the RIGHT of label
        if bx > lx:
            dy = abs(by - ly)
            if dy < min_dist:
                min_dist = dy
                best_box = (bx, by, bw, bh)

    if best_box:
        fields.append({
            "label": label["text"],
            "box": best_box
        })

# ───────────── DRAW RESULTS ─────────────
for f in fields:
    bx, by, bw, bh = f["box"]

    # Draw input field box
    cv2.rectangle(img, (bx, by), (bx + bw, by + bh), (0, 255, 0), 2)

    # Draw label text
    cv2.putText(
        img,
        f["label"],
        (bx, by - 5),
        cv2.FONT_HERSHEY_SIMPLEX,
        0.5,
        (255, 0, 0),
        1
    )

    print(f"Field: {f['label']} → x={bx}, y={by}, w={bw}, h={bh}")

# Show output
cv2.imshow("Form Fields Detected", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import pytesseract
from pytesseract import Output

# ─────────────────────────────────────────────
# CONFIG (Windows only)
# pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"
# ─────────────────────────────────────────────

IMAGE_PATH = "CMISAOP_1.png"

# Strings to search (case-insensitive)
TARGET_STRINGS = [
    "FUNCTION CODE",
    "SOL ID",
    "CCY",
    "CIF ID",
    "SCHEME CODE",
    "PRODUCT GROUP",
    "AUTHENTICATION MODE"
]

# Load image
img = cv2.imread(IMAGE_PATH)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Improve OCR accuracy
gray = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY_INV | cv2.THRESH_OTSU)[1]
gray = cv2.bitwise_not(gray)

# OCR
data = pytesseract.image_to_data(
    gray,
    output_type=Output.DICT,
    config="--psm 6"
)

# ───────────── GROUP WORDS BY LINE ─────────────
lines = {}

n = len(data["text"])
for i in range(n):
    text = data["text"][i].strip()
    if not text:
        continue

    x, y, w, h = (
        data["left"][i],
        data["top"][i],
        data["width"][i],
        data["height"][i]
    )

    line_key = (data["block_num"][i], data["par_num"][i], data["line_num"][i])

    if line_key not in lines:
        lines[line_key] = []

    lines[line_key].append({
        "text": text,
        "x": x,
        "y": y,
        "w": w,
        "h": h
    })

# ───────────── SEARCH TARGET STRINGS ─────────────
found = []

for line in lines.values():
    line_text = " ".join(w["text"] for w in line).upper()

    for target in TARGET_STRINGS:
        if target in line_text:
            xs = [w["x"] for w in line]
            ys = [w["y"] for w in line]
            ws = [w["x"] + w["w"] for w in line]
            hs = [w["y"] + w["h"] for w in line]

            bbox = (
                min(xs),
                min(ys),
                max(ws) - min(xs),
                max(hs) - min(ys)
            )

            found.append((target, bbox))

# ───────────── DRAW RESULTS ─────────────
for text, (x, y, w, h) in found:
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)
    cv2.putText(
        img,
        text,
        (x, y - 5),
        cv2.FONT_HERSHEY_SIMPLEX,
        0.5,
        (0, 0, 255),
        1
    )
    print(f"FOUND: {text} at x={x}, y={y}, w={w}, h={h}")

# Show result
cv2.imshow("Detected Strings", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import pytesseract
from pytesseract import Output
import re

IMAGE_PATH = "CMISAOP_1.png"

TARGET_STRINGS = [
    "FUNCTION CODE",
    "SOL ID",
    "CCY",
    "CIF ID",
    "SCHEME CODE",
    "PRODUCT GROUP",
    "AUTHENTICATION MODE"
]

def normalize(text):
    text = text.upper()
    text = re.sub(r'[^A-Z0-9]', '', text)
    text = text.replace('0', 'O')
    return text

img = cv2.imread(IMAGE_PATH)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

data = pytesseract.image_to_data(
    gray,
    output_type=Output.DICT,
    config="--psm 6"
)

# Collect words
words = []
for i in range(len(data["text"])):
    t = data["text"][i].strip()
    if not t:
        continue

    words.append({
        "raw": t,
        "norm": normalize(t),
        "x": data["left"][i],
        "y": data["top"][i],
        "w": data["width"][i],
        "h": data["height"][i]
    })

results = {}

for target in TARGET_STRINGS:
    target_norm = normalize(target)

    for i in range(len(words)):
        merged_text = ""
        xs, ys, ws, hs = [], [], [], []

        for j in range(i, min(i + 5, len(words))):
            merged_text += words[j]["norm"]
            xs.append(words[j]["x"])
            ys.append(words[j]["y"])
            ws.append(words[j]["x"] + words[j]["w"])
            hs.append(words[j]["y"] + words[j]["h"])

            if target_norm == merged_text:
                results[target] = (
                    min(xs),
                    min(ys),
                    max(ws) - min(xs),
                    max(hs) - min(ys)
                )
                break

# Output locations only
for k, v in results.items():
    print(f"{k} -> {v}")
```
